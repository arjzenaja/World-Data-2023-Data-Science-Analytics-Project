```python
# Mengatasi missing value
for column in df.columns:
    if df[column].dtype == 'object':
        # Jika kolom bertipe object, isi dengan mode
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        # Jika kolom bertipe numerik, isi dengan mean
        df[column].fillna(df[column].mean(), inplace=True)
```

4. Cleaning Outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = ['Population', 'GDP', 'Life expectancy'] # Fixed: Changed 'Life Expectancy' to 'Life expectancy'

# Visualisasi outlier sebelum dibersihkan
for col in numeric_cols:
    plt.figure(figsize=(8, 1.5))
    sns.boxplot(x=df[col], color='orange')
    plt.title(f'Before Cleaning: {col}')
    plt.xlabel(col)
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = ['Population', 'GDP', 'Life expectancy']

# Visualisasi outlier sebelum dibersihkan
for col in numeric_cols:
    # Check if the column contains numeric data before proceeding
    if pd.api.types.is_numeric_dtype(df[col]):
        plt.figure(figsize=(8, 1.5))
        sns.boxplot(x=df[col], color='orange')
        plt.title(f'Before Cleaning: {col}')
        plt.xlabel(col)
        plt.show()
    else:
        print(f"Skipping outlier removal for column '{col}' as it's not numeric.")

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    cleaned_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return cleaned_df

# Terapkan pembersihan untuk setiap kolom numerik
for col in numeric_cols:
    # Ensure the column is numeric before applying outlier removal
    if pd.api.types.is_numeric_dtype(df[col]):
        df = remove_outliers_iqr(df, col)
```

```python
# Visualisasi setelah pembersihan outlier
for col in numeric_cols:
    plt.figure(figsize=(8, 1.5))
    sns.boxplot(x=df[col], color='green')
    plt.title(f'After Cleaning: {col}')
    plt.xlabel(col)
    plt.show()
```

5. Exploratory Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram Distribusi Populasi
plt.figure(figsize=(8, 5))
sns.histplot(df['Population'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Population')
plt.xlabel('Population')
plt.ylabel('Count')
plt.show()

# Histogram Distribusi GDP
plt.figure(figsize=(8, 5))
sns.histplot(df['GDP'], kde=True, bins=30, color='salmon')
plt.title('Distribution of GDP')
plt.xlabel('GDP')
plt.ylabel('Count')
plt.show()
```

```python
# Korelasi antar kolom numerik
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
top_gdp = df.sort_values(by='GDP', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='GDP', y='Country', data=top_gdp, palette='viridis')
plt.title('Top 10 Countries by GDP')
plt.xlabel('GDP')
plt.ylabel('Country')
plt.show()
```

```python
top_pop = df.sort_values(by='Population', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Population', y='Country', data=top_pop, palette='magma')
plt.title('Top 10 Countries by Population')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```
